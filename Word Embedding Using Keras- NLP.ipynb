```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk','the glass of juice','the cup of tea','i am good boy','i am good developer','understand the meaning of words','you video is good']
```

```python
sent
```

```python
voc_size=10000
```

One Hot Representation

```python
one_hotRepr=[one_hot(words,voc_size) for words in sent]
```

```python
one_hotRepr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hotRepr,padding='pre',maxlen=sent_length)

#embedded_docs=pad_sequences(one_hotRepr,padding='post',maxlen=sent_legth)
```

```python
embedded_docs
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
